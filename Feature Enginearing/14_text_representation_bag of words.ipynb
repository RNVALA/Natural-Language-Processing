```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('spam.csv')
df.head()
```

```python
df.Category.value_counts()#in pandas there are function called value count that will count the values
```

```python
df['spam']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.Message,df.spam,test_size=0.2,random_state=10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
```

```python
X_train_cv=v.fit_transform(X_train.values)
X_train_cv
```

```python
#now step 4 machine learning model
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_cv,y_train)
```

```python
#now i want to evaluate the performance of the model using the countvectoru=ization we will transform x_test because x_test is just email
X_test_cv=v.transform(X_test)
X_test_cv
```

```python

from sklearn.metrics import classification_report
y_pred=model.predict(X_test_cv)
classification_report=(y_test,y_pred)
#first give the truth that is y_test and then give prediction

```

```python
emails=[
    'Hey mohan can we get together to watch football game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for  you. Dont miss this reward!'
]
emails_count=v.transform(emails)
model.predict(emails_count)
```

```python
#we will do same by sklearn pipeline
from sklearn.pipeline import Pipeline
clf=Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
emails=[
    'Hey mohan can we get together to watch football game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for  you. Dont miss this reward!'
]

clf.predict(emails)
y=clf.predict(X_test)
```

```python
classification_report(y_test,y)
```
